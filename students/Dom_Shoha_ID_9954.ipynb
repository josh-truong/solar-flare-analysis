# **Task 4: Determine total energy of flare**

Import the XRS longwave data of the solar flare you chose onto your personal computer. You can follow the same procedure that you used in the Week 5 lab.

```
import pandas as pd

# You will need to change the filename
solar_data = pd.read_csv('noaa_goes15_xrs_1m.csv')     # Import data using pandas library

display(solar_data)       # Display the dataframe of the solar data
```

The solar flare data is in the units of solar irradiance ($\frac{W}{m^2}$). As you know by now, the next step is to convert to energy per second ($ergs/s$) and convert time from ISO units to seconds. 

You've done this multiple times before at this point, so we'll just re-use the kind of code you've previously written to do this: 

```
import numpy as np
import datetime as dt

# Convert df([time (yyyy-MM-dd'T'HH:mm:ss)]) into elasped time in seconds
dates = [dt.datetime.strptime(date, '%Y-%m-%dT%H:%M:%S') for date in solar_data.iloc[:,0]]
solar_data.iloc[:,0] = [(date - dates[0]).total_seconds() for date in dates]
# Rename dataframe index
solar_data = solar_data.rename(columns={'time (yyyy-MM-dd\'T\'HH:mm:ss)': 'time (seconds)'}) 

w2erg = 10**7
int_m2 = 4*np.pi*(1.496*10**11)**2 
# Convert units to ergs
solar_data['longwave (W/m^2)'] = solar_data['longwave (W/m^2)']*w2erg*int_m2  
# Rename dateframe    
solar_data = solar_data.rename(columns={'longwave (W/m^2)': 'longwave (ergs/s)'}) 

display(solar_data)
```

Next, let's plot the solar flare data as energy per second ($ergs/s$) versus time ($seconds$), so that we can clearly visualize the flare. Again, we've done this before, so we can reuse code:




```
import matplotlib.pyplot as plt

plt.plot(solar_data['time (seconds)'], solar_data['longwave (ergs/s)'])
plt.ylabel('Power (ergs/s)')
plt.xlabel('Time (s)')

plt.show()
```

## Basline correction

As you learned in [Week 7 lecture](https://www.youtube.com/watch?v=XlbRg3e4T9g&feature=youtu.be), one of the primary challenges in determining the total energy of the flare is dealing with the "background." The sun is giving off x-rays even before a flare happens, and we do not want to accidentally include the energy of those x-rays as part of the flare energy. As a result, we need to determine the amount of x-ray power being emitted before the flare (the "baseline") and subtract it from the observed power during the flare to make sure we are only determining the *new* energy created by the flare itself.

We are importing data on a solar flare's energy that is given in Wavelength per meters squared (W/m^2). We want to convert that energy reading into ergs/s in order to compile it with data from other flares. First we take the energy in W/m^2 and apply our conversion factors to switch it into terms of ergs/s. 

We also get the time variable of our data in ISO format which we have to adjust into arbitrary units of time to plot our flare data.

w2erg = 10^7  
int_m2 = $\pi$(1.496 * 10^11)^2       
solar_data['longwave (W/m^2)'] = solar_data['longwave (W/m^2)'] * w2erg * int_m2

multiplies the flare's energy by the conversion factor needed.

One of the unique challenges we faced was determining the start of the flare. We used python  stdbe = np.std(solar_data[0:75]) code to determine the standard deviation of the mean. After we saw an energy increase above one standard deviation we considered that to be statistically significant. This limit was triggered at 75 arbitrary time units. 
The method we used to determine the termination of the flare was analyzing the downward trend for anomalous behavior that would indicate a second flare. We clearly saw a downward trend interrupted by a sharp increase in energy gain. We used this as our termination point.




```
#Import the integrate function; we will use trapezoidal integradion ('integreat.trapz')
import scipy.integrate as integrate

## Write your code here

# Calculating area under the curve without baseline correction
A_no_correction = integrate.trapz(solar_data.iloc[75:179,1], solar_data.iloc[75:179,0]) 

# Calculating area under the curve with baseline correction
A_with_correction = integrate.trapz(solar_data.iloc[75:179,1], solar_data.iloc[75:179,0]) - (avg_baseline * (179-75) * 60) 

#calculating the % error between no baseline correction and with baseline correction
err = ((A_no_correction - A_with_correction)/A_with_correction)*100
print('\n Total area under the curve without baseline correction: ', A_no_correction, ' ergs')
print('\n Total area under the curve with baseline correction: ', A_with_correction, ' ergs')
print('\n The percent difference in areas under the curves with and without baseline correction: ', err, ' %')
```
