# **Task 4: Determine total energy of flare**

Import the XRS longwave data of the solar flare you chose onto your personal computer. You can follow the same procedure that you used in the Week 5 lab.

```
import pandas as pd

# You will need to change the filename
solar_data = pd.read_csv('noaa_goes15_xrs_1m.csv')     # Import data using pandas library

display(solar_data)       # Display the dataframe of the solar data
```

The solar flare data is in the units of solar irradiance ($\frac{W}{m^2}$). As you know by now, the next step is to convert to energy per second ($ergs/s$) and convert time from ISO units to seconds. 

You've done this multiple times before at this point, so we'll just re-use the kind of code you've previously written to do this: 

```
import numpy as np
import datetime as dt

# Convert df([time (yyyy-MM-dd'T'HH:mm:ss)]) into elasped time in seconds
dates = [dt.datetime.strptime(date, '%Y-%m-%dT%H:%M:%S') for date in solar_data.iloc[:,0]]
solar_data.iloc[:,0] = [(date - dates[0]).total_seconds() for date in dates]
# Rename dataframe index
solar_data = solar_data.rename(columns={'time (yyyy-MM-dd\'T\'HH:mm:ss)': 'time (seconds)'}) 

w2erg = 10**7
int_m2 = 4*np.pi*(1.496*10**11)**2 
# Convert units to ergs
solar_data['longwave (W/m^2)'] = solar_data['longwave (W/m^2)']*w2erg*int_m2  
# Rename dateframe    
solar_data = solar_data.rename(columns={'longwave (W/m^2)': 'longwave (ergs/s)'}) 

display(solar_data)
```

Next, let's plot the solar flare data as energy per second ($ergs/s$) versus time ($seconds$), so that we can clearly visualize the flare. Again, we've done this before, so we can reuse code:




```
import matplotlib.pyplot as plt

plt.plot(solar_data['time (seconds)'], solar_data['longwave (ergs/s)'])
plt.ylabel('Power (ergs/s)')
plt.xlabel('Time (s)')

plt.show()
```

## Basline correction

As you learned in [Week 7 lecture](https://www.youtube.com/watch?v=XlbRg3e4T9g&feature=youtu.be), one of the primary challenges in determining the total energy of the flare is dealing with the "background." The sun is giving off x-rays even before a flare happens, and we do not want to accidentally include the energy of those x-rays as part of the flare energy. As a result, we need to determine the amount of x-ray power being emitted before the flare (the "baseline") and subtract it from the observed power during the flare to make sure we are only determining the *new* energy created by the flare itself.

# **The goal is to observe the pre and post flare for finding the means to create a baseline. The method we chose to eliminate preflare data was to use the threshold method. We chose this method to eliminate the higher peaks in the preflare and to get a general average. Our preflare data had lots of peaks and valleys but consistently stayed on a linear horizontal trend.**

```


## Write your code here
# You may need the statistics library for determining means and medians
import statistics as stats     

# You will need to plot your flare (as shown in the examples of baseline corrections notebook)
# in order to look at the pre-flare data and show the baseline corrected flare
import matplotlib.pyplot as plt
plt.plot(solar_data.iloc[0:160,1])   # preflare data graphed
plt.title("Solar_Data Flare Example")
plt.ylabel("Power (ergs/s)")
plt.xlabel("Time (minutes)")
plt.show()      




plt.plot(solar_data.iloc[210:325,1])   # data after flare
plt.title("Solar_Data Flare Example")
plt.ylabel("Power (ergs/s)")
plt.xlabel("Time (min)")
plt.show()

import matplotlib.pyplot as plt

plt.plot(solar_data.iloc[0:325,1])
plt.title("C-Class Flare Example")    ## entire flare with pre and post flare data graphed
plt.ylabel("Power (ergs/s)")
plt.xlabel("Time (min)")
plt.show()

pre_flare_thresh = np.clip(solar_data.iloc[0:160,1], 0, 2.3*10**(24))  ##creating the threshold for the preflare data

plt.plot(pre_flare_thresh)
plt.title("First C-Class Flare")
plt.ylabel("Power (ergs/s)")
plt.xlabel("Time (arbitrary units)")
plt.show()

import statistics as stats    
import matplotlib.pyplot as plt

plt.plot(solar_data.iloc[0:160,1])   # Zoom in on preflare data and add baseline to the graph
plt.title("C-class pre-flare with baseline")
plt.ylabel("Power (ergs/s)")
plt.xlabel("Time (arbitraty units)")

avg_baseline_C9 = pre_flare_thresh.mean()  ## set the baseline = to the threshold

x_coordinates = [0, 160]
y_coordinates = [avg_baseline_C9, avg_baseline_C9]
plt.plot(x_coordinates, y_coordinates, 'r--')
plt.show()

baseline_correction_C9 = solar_data.iloc[0:300,1].subtract(avg_baseline_C9)   ## subtract the baseline from the c-class flare

plt.plot(baseline_correction_C9)
plt.show()
```

# **We will first integrate the data with the threshold/baseline subtracted, over the whole length of our dataset. Next we integrate the whole length of the data set without the baseline subtracted. This will show the difference between the two data sets. The area under the curve with the baseline subtracted should be smaller than the area without.**

```
#Import the integrate function; we will use trapezoidal integration ('integrate.trapz')
import scipy.integrate as integrate


A_C9blc = integrate.trapz(baseline_correction_C9, solar_data.iloc[0:300,0]) #integrate data with threshold subtracted



without=integrate.trapz(solar_data.iloc[0:300,1],solar_data.iloc[0:300,0]) #integrate data without threshold subtracted

print('\n Total area under the curve using threshold method: ', A_C9blc*60, ' ergs')  #print    # * by 60 to make in seconds
print('\n Total area under the curve without using threshold method: ', without*60, ' ergs')
```
